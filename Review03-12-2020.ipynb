### Betweenness

<font size="4">
The basic measurements were calculated for the network, but what is this information telling us? 
    
Betweenness, which measures the number of conversations (or shorter paths between random nodes) that can potentially pass through a network node. It indicates in some way whether that node is a distribution point between communities. In order not to be confused we will look for the criminals who were "bridges" in the way information was dispersed or distributed, that is what betweenness gives us.
     
The formal concept of betweenness $c_B(v)$ of a node $v$ is the sum of the fraction of all-pairs shortest paths that pass through $v$, see Equation (2).

$$
c_B(v) =\sum_{s,t \in V} \frac{\sigma(s, t|v)}{\sigma(s, t)}  \;\;\;\;\;\;\; (2)
$$

where $V$ is the set of nodes, $\sigma(s, t)$ is the number of shortest $(s, t)$-paths, and $\sigma(s, t|v)$ is the number of those paths passing through some node $v$ other than $s$, $t$. If $s = t$, $\sigma(s, t) = 1$, and if $v \in {s, t}, \sigma(s, t|v) = 0$ [2],[9]. 
</font>

<font size="4"> By plotting the distribution diagram for Betweenness in g-map, we can see that the relationship of the criminals is to some extent dispersed, as Betweenness increases how much a given node is among others.  </font>

### Closeness

<font size="4">
    Closeness centrality, Equation (3), of a node $u$ is the reciprocal of the sum of the shortest path distances from $u$ to all $n-1$ other nodes. Since the sum of distances depends on the number of nodes in the graph, closeness is normalized by the sum of minimum possible distances $n-1$.

$$
C(u) = \frac{n - 1}{\sum_{v=1}^{n-1} d(v, u)}, \;\;\;\;\;\;\; (3)
$$

where $d(v, u)$ is the shortest-path distance between $v$ and $u$, and $n$ is the number of nodes in the graph [4]. Being the average of the distances of the vertex to all the others, that is to say, it is the frequency with which the vertex appears in all the minimum ways of the network allows to find the form in as a node has relevance on a set of determined nodes, where the difference is that betweenness centrality is generally regarded as a measure of others’ dependence on a given node, and therefore as a measure of potential control. Closeness centrality is usually interpreted either as a measure of access efficiency or of independence from potential control by intermediaries.

```python
import numpy as np
def page_rank(G, d=0.85, tol=1e-2, max_iter=100):
    """Return the PageRank of the nodes in the graph. 

    :param dict G: the graph
    :param float d: the damping factor
    :param flat tol: tolerance to determine algorithm convergence
    :param int max_iter: max number of iterations
    """
    nodes = G.nodes()
    matrix = nx.adjacency_matrix(G, nodelist=nodes)
    out_degree = matrix.sum(axis=0)
    weight = matrix / out_degree
    N = G.number_of_nodes()
    pr = np.ones(N).reshape(N, 1) * 1./N

    # need to repeat the initial step twice
    # for matplotlib animation
    yield nodes, pr, "init"
    yield nodes, pr, "init"

    for it in range(max_iter):
        old_pr = pr[:]
        pr = d * weight.dot(pr) + (1-d)/N
        yield nodes, pr, it
        err = np.absolute(pr - old_pr).sum()
        if err < tol:
            return pr
```

# Node2Vec, similarity

<font size="4">
    
Node2vec is an impressive algorithm that has its basis in NLP, however, its main contribution is the definition of second-order random paths for the creation of neighborhoods, but why? BFS and DFS are very rigid notions of search and unfortunately in a real world network like a terrorist network you rarely find a single type of community that follows the same principle. That is why Node2vec arises as a solution to interpolate these two searches, for it they were based on the Skip-gram architecture [7] proposing an objective function [8] that had the ability to decide between conditional independence and symmetry in feature space, see Equation (7).

\begin{equation}
\underset{f}{max} \sum_{v \in V} \left [-log(Z_{u} = \sum_{v \in V} exp(f(u) \cdot f(v))) + \sum_{n_{i} \in N_{s}(u)} f(n_{i}) \cdot f(u)  \right ] \;\;\;\;\;\; (7)
\end{equation}

```python
node2vec = Node2Vec(g, dimensions=10, walk_length=5, num_walks=100, workers=4)
```

```python
model = node2vec.fit(window=10, min_count=1, batch_words=4) 
```

Node2vec receives 2 parameters $p$ and $q$, these parameters control the probability of visiting nodes close to the analyzed source node and explore in a deeper way, based on the Skip-gram architecture which in the case of the terrorism network obtains a node and given the neighborhoods generated by it, it has the capacity to classify the nodes that represent its context, that's why when analyzing Mohamed Atta's node we obtain that the nodes with more similarity are the following:

Evaluating heuristics efficiency (a) We will use function roc auc score from package scikit-learn .
